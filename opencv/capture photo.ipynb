```python
import cv2 as cv
```

```python
import matplotlib.pyplot as plt
```

```python
cap=cv.VideoCapture(0)
if cap.isOpened():
    ret,frame=cap.read()
    print(ret)
    print(frame)
else:
    ret=False
img1=cv.cvtColor(frame,cv.COLOR_BGR2RGB)
plt.imshow(img1)
plt.title("Camera image")
plt.xticks([])
plt.yticks([])
plt.show()
cap.release()
cv.destroyAllWindows()
```
